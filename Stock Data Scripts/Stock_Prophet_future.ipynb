```python
import pandas as pd
from fbprophet import Prophet
```

```python
df = pd.read_csv("C:/Users/RAJ BATHANI/Downloads/TimeSeries Datasets/AMGN_close.csv")
```

```python
df.columns = ['ds','y']
df.head()
```

```python
df['ds'] = pd.to_datetime(df['ds'])
df.head()
```

```python
m = Prophet()
m.fit(df)
future = m.make_future_dataframe(periods=21, freq='B')
forecast = m.predict(future)
```
