```python
import pandas as pd
import numpy as np
%matplotlib inline
from sklearn.metrics import mean_squared_error
from math import sqrt
from statsmodels.tsa.arima_model import ARMA, ARIMA, ARMAResults, ARIMAResults
from pmdarima import auto_arima

df = pd.read_csv("C:/Users/RAJ BATHANI/Downloads/TimeSeries Datasets/AMGN_close.csv", index_col = ['Date'], parse_dates = ['Date'])
df.head()
df.info()
df = df.asfreq('b')
df.index
df.Close = df.Close.fillna(method = "ffill")
df.isna().sum()

#Forecasting Period 1 week
train = df.Close[0:df.shape[0]-7]
test = df.Close[df.shape[0]-7:]
#Choosing the parameters for ARIMA model
model = ARIMA(train,order=(2,1,2))
results = model.fit()
results.summary()
start = len(train)
end = len(train) + len(test) - 1 
predictions = results.predict(start=start, end=end, typ='levels')  #Predicting the values
rmse1 = sqrt(mean_squared_error(predictions,test))
#Plotting Graphs
test.plot(legend=True, label='Actual')
predictions.plot(legend=True, label='Predictions')


#Forecasting Period 1 month
train1 = df.Close[0:df.shape[0]-24]
test1 = df.Close[df.shape[0]-22:]
#Choosing the parameters for ARIMA model
model1 = ARIMA(train,order=(3,1,1))
results1 = model1.fit()
results1.summary()
start1 = len(train1)
end1 = len(train1) + len(test1) - 1 
predictions1 = results.predict(start=start1, end=end1, typ='levels')  #Predicting the values
rmse2 = sqrt(mean_squared_error(predictions1,test1))
#Plotting Graphs
test1.plot(legend=True, label='Actual')
predictions1.plot(legend=True, label='Predictions')


#Forecasting Period 3 months
train2 = df.Close[0:df.shape[0]-72]
test2 = df.Close[df.shape[0]-72:]
#Choosing the parameters for ARIMA model
model2 = ARIMA(train,order=(2,1,2))
results2 = model2.fit()
results2.summary()
start2 = len(train2)
end2 = len(train2) + len(test2) - 1 
predictions2 = results.predict(start=start2, end=end2, typ='levels')  #Predicting the values
rmse3 = sqrt(mean_squared_error(predictions2,test2))
#Plotting Graphs
test2.plot(legend=True, label='Actual')
predictions2.plot(legend=True, label='Predictions')


#Forecasting Period 6 months
train3 = df.Close[0:df.shape[0]-72]
test3 = df.Close[df.shape[0]-72:]
#Choosing the parameters for ARIMA model
model3 = ARIMA(train,order=(1,1,1))
results3 = model3.fit()
results3.summary()
start3 = len(train3)
end3 = len(train3) + len(test3) - 1 
predictions3 = results.predict(start=start3, end=end3, typ='levels')  #Predicting the values
rmse3 = sqrt(mean_squared_error(predictions3,test3))
#Plotting Graphs
test3.plot(legend=True, label='Actual')
predictions3.plot(legend=True, label='Predictions')
```
