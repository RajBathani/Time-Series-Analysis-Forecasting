```python
import pandas as pd
from fbprophet import Prophet
from statsmodels.tools.eval_measures import rmse

df = pd.read_csv("C:/Users/RAJ BATHANI/Downloads/TimeSeries Datasets/AMGN_close.csv")
df.columns = ['ds','y']
df.head()
df['ds'] = pd.to_datetime(df['ds'])
df.head()


#Forecasting with 60% Train and 40% Test Ratio
train = df.iloc[:1057]
test = df.iloc[1057:]
m = Prophet()
m.fit(train)
future = m.make_future_dataframe(periods=705, freq='B')
forecast = m.predict(future)
predictions = forecast.iloc[-705:]['yhat']
#Calcuating Errors
rmse(predictions,test['y'])
ax = forecast.plot(x='ds',y='yhat', label='Predictions', legend=True)
test.plot(x='ds',y='y', label='Test Data', legend=True, ax=ax)

#Forecasting with 70% Train and 30% Test Ratio
train = df.iloc[:1233]
test = df.iloc[1233:]
m = Prophet()
m.fit(train)
future = m.make_future_dataframe(periods=529, freq='B')
forecast = m.predict(future)
predictions = forecast.iloc[-529:]['yhat']
#Calcuating Errors
rmse(predictions,test['y'])
#Comparing the performance by plotting the graphs
ax = forecast.plot(x='ds',y='yhat', label='Predictions', legend=True)
test.plot(x='ds',y='y', label='Test Data', legend=True, ax=ax)


#Forecasting with 80% Train and 20% Test Ratio
train = df.iloc[:1409]
test = df.iloc[1409:]
m = Prophet()
m.fit(train)
future = m.make_future_dataframe(periods=353, freq='B')
forecast = m.predict(future)
predictions = forecast.iloc[-353:]['yhat']
#Calcuating Errors
rmse(predictions,test['y'])
ax = forecast.plot(x='ds',y='yhat', label='Predictions', legend=True)
test.plot(x='ds',y='y', label='Test Data', legend=True, ax=ax)

#Forecasting with 90% Train and 10% Test Ratio
train = df.iloc[:1585]
test = df.iloc[1585:]
m = Prophet()
m.fit(train)
future = m.make_future_dataframe(periods=177, freq='B')
forecast = m.predict(future)
predictions = forecast.iloc[-177:]['yhat']
#Calcuating Errors
rmse(predictions,test['y'])
ax = forecast.plot(x='ds',y='yhat', label='Predictions', legend=True)
test.plot(x='ds',y='y', label='Test Data', legend=True, ax=ax)
```
