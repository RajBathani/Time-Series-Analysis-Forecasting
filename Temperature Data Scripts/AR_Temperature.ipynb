```python
import pandas as pd
import numpy as np
%matplotlib inline
from statsmodels.tsa.ar_model import AR, ARResults
from sklearn.metrics import mean_squared_error
from math import sqrt



df = pd.read_csv("C:/Users/RAJ BATHANI/Downloads/TimeSeries Datasets/Temperature.csv", index_col = ['Date'], parse_dates = ['Date'])
df.index.freq = 'D'



#Forecasting with 60% train and 40% test 
train = df.tempmin[:df.shape[0]-1461]
test = df.tempmin[df.shape[0]-1461:]
model = AR(train)
ARfit = model.fit()
start = len(train)
end = len(train) + len(test) - 1 
forecasted_values = ARfit.predict(start=start,end=end)
rmse1 = sqrt(mean_squared_error(forecasted_values,test))
#Plotting Graphs
test.plot()
forecasted_values.plot(legend=True, label='Predictions')

#Forecasting with 70% train and 30% test 
train = df.tempmin[:df.shape[0]-1096]
test = df.tempmin[df.shape[0]-1096:]
model = AR(train)
ARfit = model.fit()
start = len(train)
end = len(train) + len(test) - 1 
forecasted_values = ARfit.predict(start=start,end=end)
rmse1 = sqrt(mean_squared_error(forecasted_values,test))
#Plotting Graphs
test.plot()
forecasted_values.plot(legend=True, label='Predictions')

#Forecasting with 80% train and 20% test 
train = df.tempmin[:df.shape[0]-731]
test = df.tempmin[df.shape[0]-731:]
model = AR(train)
ARfit = model.fit()
start = len(train)
end = len(train) + len(test) - 1 
forecasted_values = ARfit.predict(start=start,end=end)
rmse1 = sqrt(mean_squared_error(forecasted_values,test))
#Plotting Graphs
test.plot()
forecasted_values.plot(legend=True, label='Predictions')

#Forecasting with 90% train and 10% test(1 year) 
train = df.tempmin[:df.shape[0]-366]
test = df.tempmin[df.shape[0]-366:]
model = AR(train)
ARfit = model.fit()
start = len(train)
end = len(train) + len(test) - 1 
forecasted_values = ARfit.predict(start=start,end=end)
rmse1 = sqrt(mean_squared_error(forecasted_values,test))
#Plotting Graphs
test.plot()
forecasted_values.plot(legend=True, label='Predictions')

#Forecasting with 99% train and 1% test (1 month)
train = df.tempmin[:df.shape[0]-30]
test = df.tempmin[df.shape[0]-30:]
model = AR(train)
ARfit = model.fit()
start = len(train)
end = len(train) + len(test) - 1 
forecasted_values = ARfit.predict(start=start,end=end)
rmse1 = sqrt(mean_squared_error(forecasted_values,test))
#Plotting Graphs
test.plot()
forecasted_values.plot(legend=True, label='Predictions')

#Forecasting with 99% train and 1% test (1 week)
train = df.tempmin[:df.shape[0]-7]
test = df.tempmin[df.shape[0]-7:]
model = AR(train)
ARfit = model.fit()
start = len(train)
end = len(train) + len(test) - 1 
forecasted_values = ARfit.predict(start=start,end=end)
rmse1 = sqrt(mean_squared_error(forecasted_values,test))
#Plotting Graphs
test.plot()
forecasted_values.plot(legend=True, label='Predictions')
```
