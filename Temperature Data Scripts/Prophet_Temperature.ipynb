```python
import pandas as pd
from fbprophet import Prophet
from statsmodels.tools.eval_measures import rmse

df = pd.read_csv("C:/Users/RAJ BATHANI/Downloads/TimeSeries Datasets/Temperature.csv", index_col = ['Date'], parse_dates = ['Date'])
df.index.freq = 'D'
df.columns = ['ds','y']
df.head()
df['ds'] = pd.to_datetime(df['ds'])
df.head()

#Forecasting with 1 month time period
train = df.iloc[:3622]
test = df.iloc[3622:]
m = Prophet()
m.fit(train)
future = m.make_future_dataframe(periods=30)
forecast = m.predict(future)
predictions = forecast.iloc[-30:]['yhat']
#Calcuating Errors
rmse(predictions,test['y'])
ax = forecast.plot(x='ds',y='yhat', label='Predictions', legend=True)
test.plot(x='ds',y='y', label='Test Data', legend=True, ax=ax)

#Forecasting with 3 months time period
train = df.iloc[:3561]
test = df.iloc[3561:]
m = Prophet()
m.fit(train)
future = m.make_future_dataframe(periods=91)
forecast = m.predict(future)
predictions = forecast.iloc[-91:]['yhat']
#Calcuating Errors
rmse(predictions,test['y'])
ax = forecast.plot(x='ds',y='yhat', label='Predictions', legend=True)
test.plot(x='ds',y='y', label='Test Data', legend=True, ax=ax)

#Forecasting with 6 months time period
train = df.iloc[:3470]
test = df.iloc[3470:]
m = Prophet()
m.fit(train)
future = m.make_future_dataframe(periods=182)
forecast = m.predict(future)
predictions = forecast.iloc[-182:]['yhat']
#Calcuating Errors
rmse(predictions,test['y'])
ax = forecast.plot(x='ds',y='yhat', label='Predictions', legend=True)
test.plot(x='ds',y='y', label='Test Data', legend=True, ax=ax)

#Forecasting with 1 year time period
train = df.iloc[:3287]
test = df.iloc[3287:]
m = Prophet()
m.fit(train)
future = m.make_future_dataframe(periods=365)
forecast = m.predict(future)
predictions = forecast.iloc[-365:]['yhat']
#Calcuating Errors
rmse(predictions,test['y'])
ax = forecast.plot(x='ds',y='yhat', label='Predictions', legend=True)
test.plot(x='ds',y='y', label='Test Data', legend=True, ax=ax)

```
