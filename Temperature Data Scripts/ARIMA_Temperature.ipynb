```python
import pandas as pd
import numpy as np
%matplotlib inline
from sklearn.metrics import mean_squared_error
from math import sqrt
from statsmodels.tsa.arima_model import ARMA, ARIMA, ARMAResults, ARIMAResults
from pmdarima import auto_arima


df = pd.read_csv("C:/Users/RAJ BATHANI/Downloads/TimeSeries Datasets/Temperature.csv", index_col = ['Date'], parse_dates = ['Date'])
df.index.freq = 'D'

#Forecasting Period 1 week
train = df.Close[0:df.shape[0]-7]
test = df.Close[df.shape[0]-7:]
#Choosing the parameters for ARIMA model
model = ARIMA(train,order=(3,0,1))
results = model.fit()
results.summary()
start = len(train)
end = len(train) + len(test) - 1 
predictions = results.predict(start=start, end=end, typ='levels')  #Predicting the values
rmse1 = sqrt(mean_squared_error(predictions,test))
#Plotting Graphs
test.plot(legend=True, label='Actual')
predictions.plot(legend=True, label='Predictions')

#Forecasting Period 1 month
train = df.Close[0:df.shape[0]-30]
test = df.Close[df.shape[0]-30:]
#Choosing the parameters for ARIMA model
model = ARIMA(train,order=(3,0,1))
results = model.fit()
results.summary()
start = len(train)
end = len(train) + len(test) - 1 
predictions = results.predict(start=start, end=end, typ='levels')  #Predicting the values
rmse1 = sqrt(mean_squared_error(predictions,test))
#Plotting Graphs
test.plot(legend=True, label='Actual')
predictions.plot(legend=True, label='Predictions')

#Forecasting Period 3 months
train = df.Close[0:df.shape[0]-90]
test = df.Close[df.shape[0]-90:]
#Choosing the parameters for ARIMA model
model = ARIMA(train,order=(1,0,3))
results = model.fit()
results.summary()
start = len(train)
end = len(train) + len(test) - 1 
predictions = results.predict(start=start, end=end, typ='levels')  #Predicting the values
rmse1 = sqrt(mean_squared_error(predictions,test))
#Plotting Graphs
test.plot(legend=True, label='Actual')
predictions.plot(legend=True, label='Predictions')

#Forecasting Period 6 months
train = df.Close[0:df.shape[0]-180]
test = df.Close[df.shape[0]-180:]
#Choosing the parameters for ARIMA model
model = ARIMA(train,order=(3,0,1))
results = model.fit()
results.summary()
start = len(train)
end = len(train) + len(test) - 1 
predictions = results.predict(start=start, end=end, typ='levels')  #Predicting the values
rmse1 = sqrt(mean_squared_error(predictions,test))
#Plotting Graphs
test.plot(legend=True, label='Actual')
predictions.plot(legend=True, label='Predictions')
```
