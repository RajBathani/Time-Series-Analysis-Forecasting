```python
import pandas as pd
import numpy as np
%matplotlib inline
from statsmodels.tsa.ar_model import AR, ARResults
from sklearn.metrics import mean_squared_error
from math import sqrt

df = pd.read_csv("C:/Users/RAJ BATHANI/Downloads/TimeSeries Datasets/Temperature.csv", index_col = ['Date'], parse_dates = ['Date'])
df.index.freq = 'D'
df.head()

#Calaculating Residuals
df['t'] = df['Close'].shift(1)
df['residuals'] = df['tempmin'] - df['t']

#Forecasting Period 1 week
#First Applying AR model on Actual Value
train = df.tempmin[0:df.shape[0]-7]
test = df.Close[df.shape[0]-7:]
model = AR(train)
ARfit = model.fit()
start = len(train)
end = len(train) + len(test) - 1 
forecasted_values = ARfit.predict(start=start,end=end)
#Applying AR model no Residuals
train = df.residuals[1:df.shape[0]-7]
test = df.residuals[df.shape[0]-7:]
model = AR(train)
model_fit = model.fit()
pred_residuals = model_fit.predict(start=len(train), end=len(train)+len(test)-1)
#Calculating Final Predictions
final_pred = forecasted_values+pred_residuals
rmse1 = sqrt(mean_squared_error(final_pred,test))
#Plotting Graphs
test.plot()
final_pred.plot(legend=True, label='Predictions')

#Forecasting Period 1 month
#First Applying AR model on Actual Value
train = df.Close[0:df.shape[0]-30]
test = df.Close[df.shape[0]-30:]
model = AR(train)
ARfit = model.fit()
start = len(train)
end = len(train) + len(test) - 1 
forecasted_values = ARfit.predict(start=start,end=end)
#Applying AR model no Residuals
train = df.residuals[1:df.shape[0]-30]
test = df.residuals[df.shape[0]-30:]
model = AR(train)
model_fit = model.fit()
pred_residuals = model_fit.predict(start=len(train), end=len(train)+len(test)-1)
#Calculating Final Predictions
final_pred = forecasted_values+pred_residuals
rmse1 = sqrt(mean_squared_error(final_pred,test))
#Plotting Graphs
test.plot()
final_pred.plot(legend=True, label='Predictions')

#Forecasting Period 3 months
#First Applying AR model on Actual Value
train = df.Close[0:df.shape[0]-90]
test = df.Close[df.shape[0]-90:]
model = AR(train)
ARfit = model.fit()
start = len(train)
end = len(train) + len(test) - 1 
forecasted_values = ARfit.predict(start=start,end=end)
#Applying AR model no Residuals
train = df.residuals[1:df.shape[0]-90]
test = df.residuals[df.shape[0]-90:]
model = AR(train)
model_fit = model.fit()
pred_residuals = model_fit.predict(start=len(train), end=len(train)+len(test)-1)
#Calculating Final Predictions
final_pred = forecasted_values+pred_residuals
rmse1 = sqrt(mean_squared_error(final_pred,test))
#Plotting Graphs
test.plot()
final_pred.plot(legend=True, label='Predictions')

#Forecasting Period 6 months
#First Applying AR model on Actual Value
train = df.Close[0:df.shape[0]-180]
test = df.Close[df.shape[0]-180:]
model = AR(train)
ARfit = model.fit()
start = len(train)
end = len(train) + len(test) - 1 
forecasted_values = ARfit.predict(start=start,end=end)
#Applying AR model no Residuals
train = df.residuals[1:df.shape[0]-180]
test = df.residuals[df.shape[0]-180:]
model = AR(train)
model_fit = model.fit()
pred_residuals = model_fit.predict(start=len(train), end=len(train)+len(test)-1)
#Calculating Final Predictions
final_pred = forecasted_values+pred_residuals
rmse1 = sqrt(mean_squared_error(final_pred,test))
#Plotting Graphs
test.plot()
final_pred.plot(legend=True, label='Predictions')
```
